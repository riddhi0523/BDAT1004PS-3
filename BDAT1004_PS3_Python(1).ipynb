```python
users = pd.read_csv("https://raw.githubusercontent.com/justmarkham/DAT8/master/data/u.user", sep='|')
users.head()
```

Step 4. Discover what is the mean age per occupation 

```python
mean_age = users.groupby('occupation').mean()["age"]
mean_age
```

Step 5. Discover the Male ratio per occupation and sort it from the most to the least 

Step 6. For each occupation, calculate the minimum and maximum ages 

```python
maximum_age = users.groupby('occupation').max()["age"]
print("Maximum Age:", maximum_age)
print("\n\n")
minimum_age = users.groupby('occupation').min()["age"]
print("Minimum Age:", minimum_age)
```

Step 7. For each combination of occupation and sex, calculate the mean age

```python
mean_age = users.groupby(['occupation','gender']).mean()["age"]
print(mean_age)
```

Step 8. For each occupation present the percentage of women and men

```python
sumGender = users.groupby('occupation').count()["gender"]
sumMale = users[users['gender'] == 'M'].groupby('occupation').count()['gender']
sumFemale = users[users['gender'] == 'F'].groupby('occupation').count()['gender']
perMalegender = (sumMale/sumGender) * 100
perFemalegender = (sumFemale/sumGender) * 100

print(perMalegender, perFemalegender)
```

# Question 6

Create a line plot showing the number of marriages and divorces per capita in the U.S. between 1867 and 2014. Label both lines and show the legend.
Don't forget to label your axes!

```python
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv('us-marriages-divorces-1867-2014.csv')
data.head()

#plt.figure()
data.plot(x = 'Year', y = ['Marriages_per_1000','Divorces_per_1000'], label = ['Marraiges' , 'Divorces'])
          

plt.title('Marraiges & Divorces Per Capita in USA (1867 - 2014)', color = 'Green', fontsize = 22)
plt.xlabel('Year', color = 'Gray', fontsize = 17)
plt.ylabel('Capita', color = 'Gray', fontsize = 17)

plt.show()
```
